# Changing the scale
In this exercise, we'll continue to look at the dataset containing responses from a survey of young people. Does the percentage of people reporting that they feel lonely vary depending on how many siblings they have? Let's find out using a bar plot, while also exploring Seaborn's four different plot scales ("contexts").

We've already imported Seaborn as sns and matplotlib.pyplot as plt.

```python
# Import pandas as pd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
path=r'/media/documentos/Cursos/Data Science/Python/Data_Science_Python/data_sets/'
img_path=r'/media/documentos/Cursos/Data Science/Python/Data_Science_Python/Img_saves/'
# Read the data from file using read_csv
csv_filepath=path+'young-people-survey-responses.csv'
survey_data=pd.read_csv(csv_filepath,index_col=0)
survey_data['Parents Advice'] = np.where(survey_data["Parents' advice"]==1.0,'Never',
                                np.where(survey_data["Parents' advice"]==2.0,'Rarely',
                                np.where(survey_data["Parents' advice"]==3.0,'Sometimes',
                                np.where(survey_data["Parents' advice"]==4.0,'Often','Always'))))

survey_data['Number of Siblings'] = np.where(survey_data["Siblings"]==0.0,'0',
                                    np.where(survey_data["Siblings"]==1.0,'1 - 2',
                                    np.where(survey_data["Siblings"]==2.0 ,'1 - 2',
                                    np.where(survey_data["Siblings"] > 2.0,'+3','XD'))))


survey_data["Feels Lonely"] = np.where(survey_data["Loneliness"]>2,True,False)

#student_data.rename(columns={'schoolsup':'internet'},inplace = True)
print('survey_data \n',survey_data.head(3),'\n')
print('survey_data \n',survey_data.keys(),'\n')
```

Set the scale ("context") to "paper", which is the smallest of the scale options.

```python
# Set the context to "paper"
sns.set_context("paper")

# Create bar plot
sns.catplot(x="Number of Siblings", y="Feels Lonely",
            data=survey_data, kind="bar")

# Show plot
plt.show()
```

* Change the context to `"notebook"` to increase the scale.

```python
# Set the context to "paper"
sns.set_context("notebook")

# Create bar plot
sns.catplot(x="Number of Siblings", y="Feels Lonely",
            data=survey_data, kind="bar")

# Show plot
plt.show()
```

* Change the context to `"talk"` to increase the scale.

```python
# Change the context to "talk"
sns.set_context("talk")

# Create bar plot
sns.catplot(x="Number of Siblings", y="Feels Lonely",data=survey_data, kind="bar")

# Show plot
plt.show()
```

* Change the context to `"poster"`, which is the largest scale available.

```python
# Change the context to "poster"
sns.set_context("poster")

# Create bar plot
sns.catplot(x="Number of Siblings", y="Feels Lonely",data=survey_data, kind="bar")

# Show plot
plt.show()
```
